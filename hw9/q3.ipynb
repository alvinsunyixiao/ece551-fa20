```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.linspace(-np.pi, np.pi, 1000)
```

```python
w0 = np.pi / 2
C = 1
mu = .4
H = C**2 * mu * ((1 - np.cos(w0) * np.exp(-1j*w)) / (1 - 2*np.cos(w0)*np.exp(-1j*w) + np.exp(-2j*w)) - 1)
```

```python
plt.plot(w, np.abs(1. / (1. + H)))
```

```python
H = (1 - 2*np.cos(w0)*np.exp(-1j*w) + np.exp(-2j*w)) / (1 - np.cos(w0)*np.exp(-1j*w))
plt.plot(np.abs(H))
```
