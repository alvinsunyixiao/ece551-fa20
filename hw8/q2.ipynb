```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
N = 3
```

```python
theta = np.linspace(-np.pi/2, np.pi/2, 500)
v = np.pi * np.sin(theta)
```

```python
mag = np.zeros(v.shape, dtype=np.complex128)
for k in range(-N//2, N//2 + 1):
    mag += np.exp(1j*k*v)
```

```python
plt.plot(theta, np.abs(np.sin(v * N / 2) / np.sin(v / 2)))
```

```python
plt.plot(theta, np.abs(1 + 2*np.cos(np.pi*np.sin(theta)) + 2*np.cos(2*np.pi*np.sin(theta)) + 2*np.cos(3*np.pi*np.sin(theta))))
```
