```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
v, w = np.meshgrid(np.linspace(-np.pi, np.pi, 400), np.linspace(-np.pi, np.pi, 400))
```

```python
v0 = np.pi / 4
w0 = np.pi / 2
```

```python
H = 1/4 + 1/4 * np.exp(1j * (w0 - w)) + 1/4 * np.exp(1j * (v0 - v)) + 1/4 * np.exp(1j * (w0 - w + v0 - v))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(v, w, np.abs(H), cmap='jet')
```

```python
plt.figure(figsize=(15,8))
plt.imshow(np.abs(H), cmap='hot', extent=[-1, 1, -1, 1], origin='lower')
plt.title(f'Space Frequency Magnitude Response @ $\\omega_0$ = {w0/np.pi}$\\pi$ $\\nu_0$ = {v0/np.pi}$\\pi$')
plt.xlabel('Spatial Phase Delay $\\nu$ ($\\pi$ radians)')
plt.ylabel('Signal Frequency $\\omega$ ($\\pi$ radians)')
plt.colorbar()
```
